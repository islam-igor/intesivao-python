- TV, Jornal e Rádio estão em milhares de reais

#### Importar a Base de dados

```python
# importar a base de dados para o Python

import pandas as pd

tabela = pd.read_csv('advertising.csv')
display(tabela)

# pela tabela, percebe-se que há algo errado, pois se investe mais do que Vende, porém foi informado que:
    # TV, Jornal e Rádio estão em milhares de reais
    # Vendas estão em milhões
# Sendo desnecessário fazer essa correção/ajuste

# visualizar a base e fazer os ajustes na base de dados
    #-->a Base de dados informada já estava ajustada, para nao perder tempo ajustndo durante a aula
```

#### Análise Exploratória
- Vamos tentar visualizar como as informações de cada item estão distribuídas
- Vamos ver a correlação entre cada um dos itens

Para a análise desse caso, temos que buscar a CORRELAÇÃO .

O que é a Correlação?

É a relação do item e seu impacto, por exemplo

investimento        Vendas        investimento        Vendas    
    10                1               10                2     
    20                2               20                1     
    30                3               30                3     
    
Primeiro caso = Correlação = 1 (impacto direto, quando DOBROU o investimento, DOBROU as vendas)
Segundo caso = Correlação = 0.2 (Quando dobra o investimento a venda caiu pela metade, quando triplicou a venda triplicou)
               A relação entre investimento e vendas, nesse segundo caso não parece ter muita relação, pois as vendas
               se comportaram de maneira mais 'aleatória', indentificando menor relação entre eles.
               
obs.: a Correlação pode ter valores negativos, indo de -1 a 1 os valores de correlação
obs.2: nesse exemplo nao há correlações negativas.
    

```python
#!pip install matplotlib
#!pip install seaborn 
#!pip install scikit-learn
```

```python
# análise exploratória -> entender como a sua base tá se comportando
'''
CORRELAÇÃO
TV -> Vendas
Rádio -> Vendas
Jornal -> Vendas
'''

#tabela.corr() --> FAZ A CORRELAÇÃO ENTRE OS VALORES
tabela.corr()
# a correlação entre TV e Vendas é observado no cruzamento entre TV x Vendas (0.901208)
```

#### Com isso, podemos partir para a preparação dos dados para treinarmos o Modelo de Machine Learning

- Separando em dados de treino e dados de teste

Para uma inteligencia artificial, a gente tem que passar informações para ela entender e treinar 
 depois que ela tá treinada, deve-se testá-la.

Vamos então pegar a maior parte da base de dados e alimentar como treino e retirar uma pequena parte
 da base de dados, para utilizar como teste, e verificar se os valores calculados são semelhantes aos
  coletados.
  
# IMPORTANTE :
    Deve-se informar o que deve ser aprendido, e o que deve ser utilizado para aprender.
    Queremos que encontre o valor de venda, então ele será o nosso 'Y'
    Utilizaremos/Forneceremos os dados de TV, Rádio e Jornal para serem utilizados, então ele será nosso 'X'
    
        # Y é quem você quer Prever
        # X é quem você vai usar pra Prever
    

```python
y = tabela['Vendas']
x = tabela[['TV', 'Radio', 'Jornal']] # 2 colchetes pois estou selecioando uma lista de colunas

from sklearn.model_selection import train_test_split
    #esse cara separa os dados
    
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y)
    # Esse comando vai dividir a base de dados em 4 partes.
    # Quando não identificado a divisão , ele utiliza o padrão de 25% para teste
    # para mudar utiliza-se test_size=
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
    # random_state=1 -> sempre a tabela é dividida de forma aleatória, para evitar esses casos
        #utiliza-se random_state=1
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=1)
```

#### Temos um problema de regressão - Vamos escolher os modelos que vamos usar:

- Regressão Linear
- RandomForest (Árvore de Decisão)

```python
# criar a inteligecia artificial para fazer as previsões
from sklearn.linear_model import LinearRegression 
from sklearn.ensemble import RandomForestRegressor

modelo_regressaolinear = LinearRegression()
modelo_arvoredecisao = RandomForestRegressor()

# e assim criou-se 2 inteligencias artificiais !!


modelo_regressaolinear.fit(x_treino, y_treino)  #treinando
modelo_arvoredecisao.fit(x_treino, y_treino) 
```

Foi utilizado 2 modelos de IA e treinados eles com os mesmos dados, para posteriormente serem testados e ver qual das IA's tiveram uma melhora capacidade de previsão, conforme os dados de TESTE.

#### Teste da AI e Avaliação do Melhor Modelo

- Vamos usar o R² -> diz o % que o nosso modelo consegue explicar o que acontece

```python
# Testar qual IA se aproxima melhor

previsao_regressaolinear = modelo_regressaolinear.predict(x_teste)
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)

from sklearn.metrics import r2_score 
# Vai dar a porcentagem de comparação da previsão com o resultado TESTE

print(r2_score(y_teste, previsao_regressaolinear))
print(r2_score(y_teste, previsao_arvoredecisao))
```

PARA A PROPORÇÃO DE 70% TREINO E 30% TESTE

a IA de regressão teve 90.7% de previsão
a IA de Arvpore de Decisão teve 96.5%


Para esse desafio, a Arvore de decisão tem melhor eficiência!

#### Visualização Gráfica das Previsões

```python
tabela_auxiliar = pd.DataFrame()    #Cria tabela vazia
tabela_auxiliar["y_teste"] = y_teste #criou a coluna de teste
tabela_auxiliar["Previsoes ArvoreDecisao"] = previsao_arvoredecisao #criou a coluna de previsao arvore de decisao
tabela_auxiliar["Previsoes Regressao Linear"] = previsao_regressaolinear #cria a coluna de previsao de regressao

plt.figure(figsize=(15,6))  #aumenta o tamanho do grafico
sns.lineplot(data=tabela_auxiliar)  #monta o gráfico com base na tabela
plt.show() #cria o grafico
```
